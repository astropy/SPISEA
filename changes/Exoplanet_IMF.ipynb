## Plotting Studied Power-Law Functions For Exoplanets

#### Importing Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

#### Creating the general power-law function for small masses

```python
#define function
def power_law(M, a):
    return M**(-1*a)

# M is a defined mass range according to a specified paper, and a is the associated alpha value
```

#### Defining different mass/alpha values based on published papers

***Masses are in solar masses!***

```python
#Moraux 2007 (0.03 - 0.6) a = 0.69 ± 0.15
M_mass = np.linspace(0.03, 0.6, 1000)
M_a = 0.69
M_aerr = 0.15

#Suárez 2009 (0.02 - 0.50) a = 0.60 ± 0.13
S_mass = np.linspace(0.02, 0.5, 1000)
S_a = 0.60
S_aerr = 0.13

#Lodieu 2009 (0.01 - 0.5) a = 0.5 ± 0.2
L_mass = np.linspace(0.01, 0.5, 1000)
L_a = 0.5
L_aerr = 0.2

#Béjar 2011 (0.013 - 0.1) a = 0.7 ± 0.3
B_mass = np.linspace(0.013, 0.1, 1000)
B_a = 0.7
B_aerr = 0.3

#Peña Ramírez 2012 (0.006 - 0.25) a = 0.6 ± 0.2
P_mass = np.linspace(0.006, 0.25, 1000)
P_a = 0.6
P_aerr = 0.2
```

#### Plotting all of the curves together (w/o accounting for error)

```python
#plot all of the power law functions based on mass ranges
plt.plot(M_mass, power_law(M_mass, M_a), color='r', label='Moraux (a=0.69)')
plt.plot(S_mass, power_law(S_mass, S_a), color='b', label='Suárez (a=0.60)')
plt.plot(L_mass, power_law(L_mass, L_a), color='m', label='Lodieu (a=0.50)')
plt.plot(B_mass, power_law(B_mass, B_a), color='c', label='Béjar (a=0.70)')
plt.plot(P_mass, power_law(P_mass, P_a), color='y', label='Peña Ramírez (a=0.60)')
plt.title("Power Law Mass Functions for Planetary Masses According to Published Papers")
plt.xlabel("Mass of Objects (Solar Masses)")
plt.ylabel("M^(-a)")
plt.legend()
plt.show()
```

#### Accounting for error in a

```python
#Moraux
M_aup = M_a + M_aerr
M_alow = M_a - M_aerr
M_outup = power_law(M_mass, M_aup)
M_outlow = power_law(M_mass, M_alow)

#Suárez
S_aup = S_a + S_aerr
S_alow = S_a - S_aerr
S_outup = power_law(S_mass, S_aup)
S_outlow = power_law(S_mass, S_alow)

#Lodieu
L_aup = L_a + L_aerr
L_alow = L_a - L_aerr
L_outup = power_law(L_mass, L_aup)
L_outlow = power_law(L_mass, L_alow)

#Béjar
B_aup = B_a + B_aerr
B_alow = B_a - B_aerr
B_outup = power_law(B_mass, B_aup)
B_outlow = power_law(B_mass, B_alow)

#Peña Ramírez
P_aup = P_a + P_aerr
P_alow = P_a - P_aerr
P_outup = power_law(P_mass, P_aup)
P_outlow = power_law(P_mass, P_alow)
```

#### Graphing Individual Power-Law Functions (Accounting for Error in a)

```python
#Setting up subplots
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(10,10))
plt.subplots_adjust(hspace=0.75, wspace=1.25)
fig.delaxes(ax[2, 1])

#Moraux
ax[0,0].plot(M_mass, power_law(M_mass, M_a), color='r')
ax[0,0].fill_between(M_mass, M_outup, M_outlow, color='r', alpha=0.2, label="error in a")
ax[0,0].set_title("Moraux Mass Function for Planetary Masses in Blanco 1")
ax[0,0].set_xlabel("Mass of Objects (Solar Masses)")
ax[0,0].set_ylabel("M^(-0.69 ± 0.15)")
ax[0,0].legend()

#Suárez
ax[0,1].plot(S_mass, power_law(S_mass, S_a), color='b')
ax[0,1].fill_between(S_mass, S_outup, S_outlow, color='b', alpha=0.2, label="error in a")
ax[0,1].set_title("Suárez Mass Function for Planetary Masses in 25 Ori Group")
ax[0,1].set_xlabel("Mass of Objects (Solar Masses)")
ax[0,1].set_ylabel("M^(-0.60 ± 0.13)")
ax[0,1].legend()

#Lodieu
ax[1,0].plot(L_mass, power_law(L_mass, L_a), color='m')
ax[1,0].fill_between(L_mass, L_outup, L_outlow, color='b', alpha=0.2, label="error in a")
ax[1,0].set_title("Lodieu Mass Function for Planetary Masses in UKIDSS Galactic Clusters")
ax[1,0].set_xlabel("Mass of Objects (Solar Masses)")
ax[1,0].set_ylabel("M^(-0.5 ± 0.2)")
ax[1,0].legend()

#Béjar
ax[1,1].plot(B_mass, power_law(B_mass, B_a), color='c')
ax[1,1].fill_between(B_mass, B_outup, B_outlow, color='c', alpha=0.2, label="error in a")
ax[1,1].set_title("Béjar Mass Function for Planetary Masses in σ Orionis")
ax[1,1].set_xlabel("Mass of Objects (Solar Masses)")
ax[1,1].set_ylabel("M^(-0.7 ± 0.3)")
ax[1,1].legend()

#Peña Ramírez
ax[2,0].plot(P_mass, power_law(P_mass, P_a), color='y')
ax[2,0].fill_between(P_mass, P_outup, P_outlow, color='y', alpha=0.2, label="error in a")
ax[2,0].set_title("Peña Ramírez Mass Function for Planetary Masses in σ Orionis")
ax[2,0].set_xlabel("Mass of Objects (Solar Masses)")
ax[2,0].set_ylabel("M^(-0.6 ± 0.2)")
ax[2,0].legend()


plt.show()
```

### Creating a General Power-Law Curve Fit

```python
#combine all mass data
all_masses = []
for a in M_mass:
    all_masses.append(a)
for b in S_mass:
    all_masses.append(b)
for c in L_mass:
    all_masses.append(c)
for d in B_mass:
    all_masses.append(d)
for e in P_mass:
    all_masses.append(e)
```

```python
print(len(all_masses)) #to make sure the correct amount of entries were appended
```

```python
#combine all power_law output data
all_power = []
for f in M_mass:
    all_power.append(power_law(f, M_a))
for g in S_mass:
    all_power.append(power_law(g, S_a))
for h in L_mass:
    all_power.append(power_law(h, L_a))
for k in B_mass:
    all_power.append(power_law(k, B_a))
for l in P_mass:
    all_power.append(power_law(l, P_a))
```

```python
print(len(all_power))
```

```python
#use scipy curve fit to create a general power law
popt, pcov = curve_fit(power_law, all_masses, all_power)   #popt gives fit value for a
avg_a = popt

#create an array with mass values using fitted a
avg_power = power_law(all_masses, avg_a)

#plot combined mass and power law datasets
plt.figure()
plt.scatter(all_masses, all_power)
plt.plot(all_masses, avg_power, color='r')
plt.show()
```
