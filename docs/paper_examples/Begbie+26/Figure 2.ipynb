## Below is the code to recreate Figure 2.

Topic: Introducing and comparing initial mass functions extending to the brown dwarf mass regime. 

```python
# Importing necessary packages.
from spisea.imf import imf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define IMFs with substellar capabilities
simf = imf.Salpeter_Kirkpatrick_2024()
oimf = imf.Weidner_Kroupa_2004()

# Normalize the IMFs
simf.normalize(Mcl=1e4)  # 10^4 Msun cluster
oimf.normalize(Mcl=1e4)

# Define a mass grid from 0.01 to 120 M_sun
m_min = 0.01
m_max = 120

m_grid = np.logspace(np.log10(m_min), np.log10(m_max), 2000)

# find the dN/dm values
s_xi_vals = simf.xi(m_grid)   # dN/dm
o_xi_vals = oimf.xi(m_grid)

s_xi_logm = m_grid * s_xi_vals
o_xi_logm = m_grid * o_xi_vals
```

```python
# Plot the figure
plt.figure(figsize=(8,6))

plt.loglog(m_grid, o_xi_vals, lw=2, color='black', label='Normalized Weidner_Kroupa_2004 IMF')
plt.loglog(m_grid, s_xi_vals, lw=2, color='red', label='Normalized Salpeter_Kirkpatrick_2024 IMF')
plt.axvspan(0.01, 0.08, alpha=0.3, color='tab:blue', label='Brown Dwarfs (M < 0.08 $M_\\odot$)')

plt.xlabel(r'Mass $(M_\odot)$', fontsize=16)
plt.ylabel(r'dN/dm', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('SPISEA Normalized Brown Dwarf IMFs', fontsize=20)
plt.legend(fontsize=14)

plt.grid(alpha=0.3, which='both')
plt.tight_layout()
#plt.savefig('imf.png')
plt.show()
```
