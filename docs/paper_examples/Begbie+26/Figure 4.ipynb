## Below is the code to recreate Figure 4.

Topic: Showing the Gaussian Process-based interpolation between the Phillips and Pisa/Parsec Evolutionary Models

```python
# Importing necessary packages.
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
# Unpacking the merged Phillips and Pisa/PARSEC files for logage 6 and 9
evo_folder = '/System/Volumes/Data/mnt/g/lu/models/evolution/merged'

philpisafile = evo_folder + '/phillips_pisa/z015/iso_6.00.fits'
philparsfile = evo_folder + '/phillips_parsec/z015/iso_9.00.fits'

philpisa = Table.read(philpisafile, format='fits')
philpars = Table.read(philparsfile, format='fits')

philpisa_m = philpisa['Mass']
philpisa_l = philpisa['L']
philpisa_t = philpisa['Teff']
philpisa_g = philpisa['logg']
philpisa_interp = np.where(philpisa['interpolated'] == True)
phil1 = philpisa['Mass'] < 0.075
pisa = philpisa['Mass'] > 0.2

philpars_m = philpars['Mass']
philpars_l = philpars['L']
philpars_t = philpars['Teff']
philpars_g = philpars['logg']
philpars_interp = np.where(philpars['interpolated'] == True)
phil2 = philpars['Mass'] <= 0.075
parsec = philpars['Mass'] >= 0.2
```

```python
# Plotting the Phillips to Pisa interpolation
fig, axs = plt.subplots(1, 3, figsize=(15,5))
plt.suptitle('Interpolating from Phillips to Pisa for a 1 Myr Cluster', fontsize=30, fontweight='bold')

axs[0].scatter(philpisa_m[phil1], philpisa_l[phil1], label='Phillips', color='red', marker='d', s=70)
axs[0].scatter(philpisa_m[pisa], philpisa_l[pisa], label='Pisa', color='blue', marker='s', s=70)
axs[0].scatter(philpisa_m[philpisa_interp], philpisa_l[philpisa_interp], label='Interpolated Values', color='black', s=100)
axs[0].axvspan(0.075, 0.2, alpha=0.3, color='tab:gray', label='Interpolation Region')
axs[0].set_title('Mass vs. Luminosity', fontsize=24)
axs[0].set_xlabel('Mass (M$_\odot$)', fontsize=20)
axs[0].set_ylabel('Luminosity', fontsize=20)
axs[0].set_xlim(0,0.4)
axs[0].set_ylim(-4, 0.5)
axs[0].tick_params(axis='both', labelsize=18)
axs[0].grid()
axs[0].legend(loc='lower right', fontsize=16)

axs[1].scatter(philpisa_m[phil1], philpisa_t[phil1], label='Phillips', color='red', marker='d', s=70)
axs[1].scatter(philpisa_m[pisa], philpisa_t[pisa], label='Pisa', color='blue', marker='s', s=70)
axs[1].scatter(philpisa_m[philpisa_interp], philpisa_t[philpisa_interp], label='Interpolated Values', color='black', s=100)
axs[1].axvspan(0.075, 0.2, alpha=0.3, color='tab:gray', label='Interpolation Region')
axs[1].set_title('Mass vs. Effective Temperature', fontsize=24)
axs[1].set_xlabel('Mass (M$_\odot$)', fontsize=20)
axs[1].set_ylabel('Teff (K)', fontsize=20)
axs[1].set_xlim(0,0.4)
axs[1].set_ylim(2.7, 3.7)
axs[1].tick_params(axis='both', labelsize=18)
axs[1].grid()
axs[1].legend(loc='lower right', fontsize=16)

axs[2].scatter(philpisa_m[phil1], philpisa_g[phil1], label='Phillips', color='red', marker='d', s=70)
axs[2].scatter(philpisa_m[pisa], philpisa_g[pisa], label='Pisa', color='blue', marker='s', s=70)
axs[2].scatter(philpisa_m[philpisa_interp], philpisa_g[philpisa_interp], label='Interpolated Values', color='black', s=100)
axs[2].axvspan(0.075, 0.2, alpha=0.3, color='tab:gray', label='Interpolation Region')
axs[2].set_title('Mass vs. Surface Gravity', fontsize=24)
axs[2].set_xlabel('Mass (M$_\odot$)', fontsize=20)
axs[2].set_ylabel('log(gravity)', fontsize=20)
axs[2].set_xlim(0,0.4)
axs[2].set_ylim(2.5, 3.7)
axs[2].tick_params(axis='both', labelsize=18)
axs[2].grid()
axs[2].legend(loc='lower right', fontsize=16)

plt.tight_layout()
#plt.savefig('philtopisa.png')
plt.show()
```

```python
# Plotting the Phillips to PARSEC interpolation
fig, axs = plt.subplots(1, 3, figsize=(15,5))
plt.suptitle('Interpolating from Phillips to PARSEC for a 1 Gyr Cluster', fontsize=30, fontweight='bold')

axs[0].scatter(philpars_m[phil2], philpars_l[phil2], label='Phillips', color='red', marker='d', s=70)
axs[0].scatter(philpars_m[parsec], philpars_l[parsec], label='PARSEC', color='orange', marker='P', s=70)
axs[0].scatter(philpars_m[philpars_interp], philpars_l[philpars_interp], label='Interpolated Values', color='black', s=100)
axs[0].axvspan(0.075, 0.2, alpha=0.3, color='tab:gray', label='Interpolation Region')
axs[0].set_title('Mass vs. Luminosity', fontsize=24)
axs[0].set_xlabel('Mass (M$_\odot$)', fontsize=20)
axs[0].set_ylabel('Luminosity', fontsize=20)
axs[0].set_xlim(0,0.4)
axs[0].set_ylim(-7, -1)
axs[0].tick_params(axis='both', labelsize=18)
axs[0].grid()
axs[0].legend(loc='lower right', fontsize=16)

axs[1].scatter(philpars_m[phil2], philpars_t[phil2], label='Phillips', color='red', marker='d', s=70)
axs[1].scatter(philpars_m[parsec], philpars_t[parsec], label='PARSEC', color='orange', marker='P', s=70)
axs[1].scatter(philpars_m[philpars_interp], philpars_t[philpars_interp], label='Interpolated Values', color='black', s=100)
axs[1].axvspan(0.075, 0.2, alpha=0.3, color='tab:gray', label='Interpolation Region')
axs[1].set_title('Mass vs. Effective Temperature', fontsize=24)
axs[1].set_xlabel('Mass (M$_\odot$)', fontsize=20)
axs[1].set_ylabel('Teff (K)', fontsize=20)
axs[1].set_xlim(0,0.4)
axs[1].set_ylim(2.4, 3.7)
axs[1].tick_params(axis='both', labelsize=18)
axs[1].grid()
axs[1].legend(loc='lower right', fontsize=16)

axs[2].scatter(philpars_m[phil2], philpars_g[phil2], label='Phillips', color='red', marker='d', s=70)
axs[2].scatter(philpars_m[parsec], philpars_g[parsec], label='PARSEC', color='orange', marker='P', s=70)
axs[2].scatter(philpars_m[philpars_interp], philpars_g[philpars_interp], label='Interpolated Values', color='black', s=100)
axs[2].axvspan(0.075, 0.2, alpha=0.3, color='tab:gray', label='Interpolation Region')
axs[2].set_title('Mass vs. Surface Gravity', fontsize=24)
axs[2].set_xlabel('Mass (M$_\odot$)', fontsize=20)
axs[2].set_ylabel('log(gravity)', fontsize=20)
axs[2].set_xlim(0,0.4)
axs[2].set_ylim(3.5, 5.5)
axs[2].tick_params(axis='both', labelsize=18)
axs[2].grid()
axs[2].legend(loc='lower right', fontsize=16)

plt.tight_layout()
#plt.savefig('philtoparsec.png')
plt.show()
```
