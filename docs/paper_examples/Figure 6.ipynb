## Below is the code to re-create Figure 6

### Topic: WR Star Populations Assuming Different Evolution Models

```python
from spisea import evolution, atmospheres, synthetic
from spisea import reddening as pop_red
from spisea.imf import imf, multiplicity
import numpy as np
import os
```

```python
red_law = pop_red.RedLawHosek18b()
atm_func = atmospheres.get_merged_atmosphere
filters =  ['acs,wfc1,f814w','wfc3,ir,f125w', 'wfc3,ir,f160w']
AKs = 0.73 # Ks filter extinction in mags
Mcl = 25000. # cluster mass in Msun
dist = 4000 # distance in parsecs

imf_min = 0.01 # minimum stellar mass in Msun
imf_max = 150. # Maximum stellar mass in Msun
imf_mass_limits = np.array([imf_min, imf_max])
imf_powers = np.array([-1.80]) # the slope of IMF, alpha
imf_multi=multiplicity.MultiplicityUnresolved()
new_imf = imf.IMF_broken_powerlaw(imf_mass_limits, imf_powers, imf_multi)

# Make 2 directories, one for each evo_model
dir1 = 'iso_merged'
dir2 = 'iso_mist'

if not os.path.exists(dir1):
    os.mkdir(dir1)
if not os.path.exists(dir2):
    os.mkdir(dir2)
```

```python
# Measure the mean and standard deviation of the number WR stars in synthetic clusters 
# as a function of logAge between 6.30 and 6.90.

evo_model = evolution.MergedBaraffePisaEkstromParsec() 

dir_iso_merged = dir1 #Direcotory for new ischrones to generate or pre-generated isochrones

mean_WR_merged = [] # a list to save the mean of the number of WR stars in each logAge bin
std_WR_merged = [] # a list to save the standard deviation of the number of WR stars in each logAge bin

for logAge in np.arange(6.30, 6.90, 0.05): # logAge from 6.30 to 6.90 with a step size of 0.05
    tmp = [] # a temporary list to save the number of WR stars in each synthetic cluster
    for i in range(0,100): # generate 100 synthetic clusters for each logAge bin
        new_iso = synthetic.IsochronePhot(logAge, AKs, dist,
                                      evo_model=evo_model, atm_func=atm_func,
                                      red_law=red_law, iso_dir=dir_iso_merged, filters=filters)
        cluster = synthetic.ResolvedCluster(new_iso, new_imf, Mcl)
        tmp.append(cluster.star_systems['isWR'].sum())
    mean_WR_merged.append(np.array(tmp).mean())
    std_WR_merged.append(np.array(tmp).std())
```

```python
evo_model = evolution.MISTv1()

dir_iso_mist = dir2

mean_WR_mist = []
std_WR_mist = [] 

for logAge in np.arange(6.30, 6.90, 0.05):
    tmp = []
    for i in range(0,100):
        new_iso = synthetic.IsochronePhot(logAge, AKs, dist,
                                      evo_model=evo_model, atm_func=atm_func,
                                      red_law=red_law, iso_dir=dir_iso_mist, filters=filters)
        cluster = synthetic.ResolvedCluster(new_iso, new_imf, Mcl)
        tmp.append(cluster.star_systems['isWR'].sum())
    mean_WR_mist.append(np.array(tmp).mean())
    std_WR_mist.append(np.array(tmp).std())
```

```python
#Make Figure

import matplotlib.pyplot as plt

plt.figure(figsize=(7, 6), dpi= 300)
plt.errorbar(np.arange(6.30, 6.90, 0.05),np.array(mean_WR_merged),yerr=std_WR_merged, label='Merged', lw=2.5,
            fmt='-o',ms=5)
plt.errorbar(np.arange(6.30, 6.90, 0.05),np.array(mean_WR_mist),yerr=std_WR_mist, label='MIST',
            fmt='-s',ms=4)
plt.xlabel('log(Age)')
plt.ylabel('WR stars')
plt.title(r"$M_{cl} = 25,000\,M_{\odot},\,\alpha=1.80$")
plt.ylim([-2,37])
plt.tick_params(direction='in')
plt.legend()
plt.savefig('WRstars.png', bbox_inches='tight')
```
